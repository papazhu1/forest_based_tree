```python
# 导入决策树分类器
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.datasets import load_iris
iris = load_iris()
# 创建决策树分类器
dtc = DecisionTreeClassifier()
# 拟合模型
dtc.fit(iris.data, iris.target)

# 输出每个节点的gini指数


# 找出决策树的叶子节点
leaves = set(dtc.apply(iris.data))
print(leaves)
tree.plot_tree(dtc)
```
